## 2. The question: decision, action, cost of a wrong call

## Research Question

Can webpages be grouped into meaningful performance archetypes using content and search metrics, so that content teams can make more informed decisions about how to manage different types of pages?

### Decision
This project helps content or SEO teams decide how different groups of webpages should be managed.

### Action
Based on the identified archetype, the team can decide whether to protect, improve, refresh, monitor, or consider pruning a page.

### Cost of a Wrong Recommendation
If pages are grouped incorrectly, the team may spend time improving pages that do not need attention or overlook pages that could benefit from optimization. This can waste resources and miss valuable opportunities.

### Why Data/ML Helps
The dataset contains many different content and search metrics. Clustering can analyze these metrics together and identify meaningful groups of similar pages that would be difficult to define using a single manual rule.
